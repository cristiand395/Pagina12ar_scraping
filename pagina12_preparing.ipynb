```python
def get_section_links(section):
    sec = requests.get(section)
    s_section = BeautifulSoup(sec.text, 'lxml')
    articles_list = s_section.find_all('article', attrs = {'class': 'article-item'})
    links_list = []
    for i in range(len(articles_list)):
        try:
            link = articles_list[i].find(attrs={'class': 'title-list'}).find('a').get('href')
            links_list.append(f'https://www.pagina12.com.ar{link}')
        except Exception as e:
            print(f'Error request: {e} \n')
    return(links_list)
```

```python
def get_sections(url): 
    web = requests.get(url)
    if web.status_code != 200:
        print(f'Error code: {web.status_code}')
    s = BeautifulSoup(web.text, 'lxml')
    sections = s.find('ul', attrs = {'class': 'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
    sections_links = [section.find('a').get('href') for section in sections]
    return sections_links
URL = 'https://www.pagina12.com.ar/'
sections = get_sections(URL)
sections
```

```python
def get_articles(sections):
    articles_list = []
    for i, section in enumerate(sections):
        req = requests.get(sections[i])
        soup = BeautifulSoup(req.text, 'lxml')
        articlees_list = soup.find_all('article', attrs = {'class': 'article-item'})
        links_list = []
        for j in range(len(articlees_list)):
            try:
                link = articlees_list[j].find(attrs={'class': 'title-list'}).find('a').get('href')
                links_list.append(f'https://www.pagina12.com.ar{link}')
            except Exception as e:
                print(f'Error request: {e} \n')
                return None
        articles_list.append(links_list)
    return articles_list
articles_sec = get_articles(sections)
articles_sec
```
